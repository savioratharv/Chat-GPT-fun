# Semantic Search

```python
text = openai.Embedding.create(
    input=list(df['text']),
    engine="text-search-babbage-query-001")
```

```python
l1=[]
for i in range(len(df)):
  l1.append(text['data'][i]['embedding'])
```

```python
from openai.embeddings_utils import get_embedding, cosine_similarity
```

```python
def search_reviews(df,search_query, n):
    embedding = get_embedding(
        search_query,
        engine="text-search-babbage-query-001"
    )
    l2=[]
    for i in l1:
      l2.append(cosine_similarity(i,embedding))
    df['similarities']=l2
    top_n =df.sort_values("similarities", ascending=False).head(n)
    return top_n
     
```

```python
from IPython.display import display, HTML
res = search_reviews(df,"Need a vegetarian cheesy potato dish which feels like both mashed potatos and french fries that can be made with 10 or less ingredients with a microwave", n=5)
display(HTML(res.to_html()))
```
